## Histogramas

```python
def hist_colors(hist):
    frec = hist[0]
    x_bins = hist[1]
    patches = hist[2]
 
    color_fracs = frec / frec.max()

    color_norm = colors.Normalize(color_fracs.min(), color_fracs.max())
    color_norm

    for frac, patch in zip(color_fracs, patches):
        color = plt.cm.Blues(color_norm(frac))
        patch.set_facecolor(color)

    return (frec, x_bins, patches)
```

```python
def histograma_config(ax_plot, x, title='', x_label='', y_label=''):
    hist_colors( ax_plot.hist(x, edgecolor='k', linewidth=1) )
    
    ax_plot.set_xlabel(x_label)
    ax_plot.set_ylabel(y_label)
    ax_plot.set_title(title)
```

```python
fig, ax = plt.subplots(3,1, True, figsize=(10,7))

hist_colors( ax[0].hist(mtcars.hp, linewidth=1, edgecolor='k') )
ax[0].set_title('Caballos de fuerza')


trash,x_bins,trash = hist_colors( ax[1].hist(mtcars.hp, linewidth=1, edgecolor='k', density=True) )

sigma = np.std(mtcars.hp)
mu = np.mean(mtcars.hp)
y = ((1 / (np.sqrt(2 * np.pi) * sigma)) *
     np.exp(-0.5 * (1 / sigma * (x_bins - mu))**2))

ax[1].plot(x_bins, y, 'k--', linewidth=1.5, label='y')
ax[1].legend()


hist_colors( ax[2].hist(mtcars.hp, linewidth=1, edgecolor='k', density=True, cumulative=True) )

y = y.cumsum()
y /= y[-1]

ax[2].plot(x_bins, y, 'k--', linewidth=1.5, label='y')
ax[2].legend()
```

```python
fig, (pib_plot, aporte_plot, internet_plot)= plt.subplots(nrows = 1, ncols = 3, figsize = (15,3), sharey = True)

histograma_config(ax_plot = pib_plot, 
                  x = orangeec['GDP PC'], 
                  title='PIB per Capita en paises latam', 
                  x_label='PIB per Capita', 
                  y_label='Cantidad de Paises')

histograma_config(aporte_plot, orangeec['Creat Ind % GDP'], 
                  'Contribucion Economia naranja al pib en paises latam', 
                  'Aporte a la Economia Naranja (%)')

histograma_config(internet_plot, orangeec['Internet penetration % population'], 
                  'Penetracion internet en paises latam', 
                  'Penetracion de internet (%)')

# fig.savefig('orangeec_hist.png')
```

## Boxplot

```python
def boxplot_config(ax, x, labels, title='', xlabel='', ylabel=''):
    boxprops = dict(linestyle='--', linewidth=2, color='darkgoldenrod')
    flierprops = dict(marker='o', markerfacecolor='green', markersize=6, label='outlier')
    medianprops = dict(linestyle='-.', linewidth=2.5, color='firebrick', label='median')
    meanpointprops = dict(marker='D', markeredgecolor='black', markerfacecolor='skyblue', label='mean')

    bbplot = ax.boxplot(x,
                    patch_artist=True,  # fill with color
                    labels=labels,
                    showmeans = True,
                    boxprops=boxprops,
                    flierprops=flierprops,
                    medianprops=medianprops,
                    meanprops=meanpointprops
                    )

    
    numero_cajas = len( bbplot['boxes'] )

    color_fracs = np.array(list( range( numero_cajas )) ) / numero_cajas

    color_norm = colors.Normalize(color_fracs.min(), color_fracs.max())
    color_norm

    for frac, patch in zip(color_fracs, bbplot['boxes']):
        color = plt.cm.summer(color_norm(frac))
        patch.set_facecolor(color)


    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.legend()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

hp_por_cilindros = [ mtcars[mtcars['cyl'] == 4]['hp'], 
                     mtcars[mtcars['cyl'] == 6]['hp'], 
                     mtcars[mtcars['cyl'] == 8]['hp'] ]

labels_cilindros = sorted(mtcars.cyl.apply(lambda x: str(x)).unique())

boxplot_config( ax[0], hp_por_cilindros, labels_cilindros,
                'HP por cilindros', 'Cilindros', 'Caballos de fuerza')



millas_trans = [ mtcars[mtcars['am'] == 0]['mpg'],
                 mtcars[mtcars['am'] == 1]['mpg'] ]

labels_trans = mtcars.am.apply(lambda x: 'Manual' if x   else 'Automatico').unique()

boxplot_config( ax[1], millas_trans, labels_trans,
                'Millas por galon - Transmision', 'Tipo transmision', 'Millas por galon')
```

```python
fig, ax = plt.subplots(2, 1, figsize=(5, 10), sharex=True)

aporte_economia = [ orangeec[orangeec['strong_economy'] == True]['Creat Ind % GDP'].dropna(),
                    orangeec[orangeec['strong_economy'] != True]['Creat Ind % GDP'].dropna() ] 

labels_economy = orangeec.strong_economy.apply(lambda x: 'Encima del promedio' if x else 'Debajo del promedio').unique()

boxplot_config( ax[0], aporte_economia, labels_economy,
                'Aporte a la economia paises con alto y baja economia', 
                ylabel='Aporte a la economia naranja')



penetracion_inter = [ orangeec[orangeec['strong_economy'] == True]['Internet penetration % population'].dropna(),
                      orangeec[orangeec['strong_economy'] != True]['Internet penetration % population'].dropna() ] 

boxplot_config( ax[1], penetracion_inter, labels_economy,
                'Penetracion de internet segun su economia', 
                ylabel='Aporte a la economia naranja',
                xlabel='Clasificacion de pais')


```
